# ДЗ


 Напишите скрипт, который преобразует входную строку в верхний регистр.
 Например, 
   'hello' -> 'HELLO'
 методы upper(), swapcase() использовать нельзя!

```python
input_string = input("Введите строку: ")
output_string = ""

for char in input_string:
    if "a" <= char <= "z":
        output_string += chr(ord(char) - 32)
    else:
        output_string += char

print("Результат: ", output_string)
```
