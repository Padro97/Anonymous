Три отличия от классической математики:

Python поддерживает работу с десятичными дробями, в то время как в классической математике это может быть недоступно.
В Python есть возможность работы со встроенными комплексными числами, тогда как в классической математике комплексные числа не используются так часто.
В Python результат деления двух целых чисел всегда является десятичной дробью, в то время как в классической математике результатом может быть целое число.
